```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("[UCI] AAAI-14 Accepted Papers - Papers.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

## since title, author, abstract can take any values, lets drop these columns


```python
df = df.drop(["title","authors","abstract"],axis=1)
```

## Searching by Regex, getting the values in brackets for groups column

```python
import re
groups = set()
for alpha in df.groups.unique():
    m = re.findall(r"\(([A-Za-z0-9_]+)\)", alpha)
    for size in m:
        groups.add(size)
```

```python
groups
```

## Creating column according to groups 
### and value = 0 , key absent in the group name
### value 1 , key present in the group name

```python
#df["AIW"]=np.where(df[df['groups'].str.contains("AIW")],1,0)
#df.loc[df['groups'].str.contains("AIW")]

for keys in groups:
    df[keys]=0
    df.loc[df['groups'].str.contains(keys), keys] = 1
```

```python
df.head().T
```

```python
from sklearn.cluster import KMeans
```

# we will keep clusters = number of keys

```python
len(groups)
```

```python
kmeans = KMeans(n_clusters=len(groups),random_state=0)
```

```python
kmeans.fit(df.iloc[:,3:])
```

```python
kmeans.cluster_centers_
```
